```python
#Data visulization
#In this visualization there are the estimated number of employees according to different Areas of India

df.columns= ["Region","Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Area"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Area", data=df)
plt.show()
```

```python
#Unemployment rate according to different regions of India

plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=df)
plt.show()
```

```python
#Exploratorary Data Analysis

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
l = ['Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate']
s = [50,50,50]
ax.pie(s,labels = l,autopct = '%1.2f%%')
plt.show()
```
